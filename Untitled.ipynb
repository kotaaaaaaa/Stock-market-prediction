```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# Generating a simple dataset
X, y = make_regression(n_samples=1000, n_features=5, noise=0.1, random_state=42)
X = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(1, 6)])
```

```python
from sklearn.ensemble import RandomForestRegressor

# Training the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X, y)

# Plotting feature importances
plt.figure(figsize=(10,6))
plt.barh(X.columns, rf_model.feature_importances_)
plt.title('Random Forest Feature Importances')
plt.show()
```

```python
import xgboost as xgb

# Training the XGBoost model
xgb_model = xgb.XGBRegressor(objective ='reg:squarederror', n_estimators=100)
xgb_model.fit(X, y)

# Plotting feature importances
plt.figure(figsize=(10,6))
plt.barh(X.columns, xgb_model.feature_importances_)
plt.title('XGBoost Feature Importances')
plt.show()
```

```python
import lightgbm as lgb

# Training the LightGBM model
lgb_model = lgb.LGBMRegressor(n_estimators=100)
lgb_model.fit(X, y)

# Plotting feature importances
plt.figure(figsize=(10,6))
plt.barh(X.columns, lgb_model.feature_importances_)
plt.title('LightGBM Feature Importances')
plt.show()
```

```python
import catboost as cb

# Training the CatBoost model
cb_model = cb.CatBoostRegressor(verbose=0, n_estimators=100)
cb_model.fit(X, y)

# Plotting feature importances
plt.figure(figsize=(10,6))
plt.barh(X.columns, cb_model.get_feature_importance())
plt.title('CatBoost Feature Importances')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier

# Load the data
data = load_breast_cancer()
X, y = data.data, data.target

# Train a Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X, y)

print(rf_model.feature_importances_)

# Plot feature importances
plt.barh(data.feature_names, rf_model.feature_importances_)
plt.xlabel('Feature Importance')
plt.ylabel('Feature Name')
plt.show()
```

```python
import xgboost as xgb

# Train an XGBoost model
xgb_model = xgb.XGBClassifier(n_estimators=100, random_state=42)
xgb_model.fit(X, y)

# Plot feature importances
plt.barh(data.feature_names, xgb_model.feature_importances_)
plt.xlabel('Feature Importance')
plt.ylabel('Feature Name')
plt.show()
```

```python
import lightgbm as lgb

# Train a LightGBM model
lgb_model = lgb.LGBMClassifier(n_estimators=100, random_state=42)
lgb_model.fit(X, y)

# Plot feature importances
plt.barh(data.feature_names, lgb_model.feature_importances_)
plt.xlabel('Feature Importance')
plt.ylabel('Feature Name')
plt.show()
```

```python
from catboost import CatBoostClassifier

# Train a CatBoost model
cat_model = CatBoostClassifier(n_estimators=100, random_seed=42, verbose=0)
cat_model.fit(X, y)

# Plot feature importances
plt.barh(data.feature_names, cat_model.get_feature_importance())
plt.xlabel('Feature Importance')
plt.ylabel('Feature Name')
plt.show()
```
